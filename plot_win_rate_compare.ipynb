```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datasets import load_dataset

# 控制是否显示误差棒
show_errorbar = False

# 载入数据
dataset = load_dataset("json", data_files="ultrafeedback_with_learning_order.json")['train']

# 模型名称
# model_names = ['llama', 'mistral', 'qwen-2.5-7b']
# model_names = ['BGE']
model_names = ['Skywork-LLaMA-3-8B']

# 目标指标
# target_metric = "embedding_distance"  # 或 "embedding_distance"
target_metric = "reward_score_diff"  # 或 "embedding_distance"

# 设置指标名称
if target_metric == "embedding_distance":
    target_metric_name = "Embedding Distance"
elif target_metric == "reward_score_diff":
    target_metric_name = "Reward Score Margin"
else:
    raise NotImplementedError

# 创建图形
plt.figure(figsize=(6, 4))
all_score_diffs = []

# 遍历每个模型
for model_name in model_names:
    # 过滤数据
    filtered_data = [
        (x["score_diff"], x[target_metric])
        for x in dataset
        if x["score_diff"] is not None and x[target_metric] is not None
    ]

    df = pd.DataFrame(filtered_data, columns=["score_diff", target_metric])
    df["score_diff"] = df["score_diff"].apply(lambda x: round(x * 2) / 2)

    # 聚合：均值和标准差
    df_grouped = df.groupby("score_diff")[target_metric].agg(["mean", "std"]).reset_index()
    all_score_diffs.extend(df_grouped["score_diff"].tolist())

    # 绘图
    if show_errorbar:
        plt.errorbar(
            df_grouped["score_diff"],
            df_grouped["mean"],
            yerr=df_grouped["std"],
            fmt='-o',
            label=model_name,
            capsize=3,
            alpha=0.8
        )
    else:
        plt.plot(
            df_grouped["score_diff"],
            df_grouped["mean"],
            marker='o',
            label=model_name,
            alpha=0.9
        )

# 设置 x 轴刻度
min_x = np.floor(min(all_score_diffs))
max_x = np.ceil(max(all_score_diffs))
plt.xticks(np.arange(min_x, max_x + 0.5, 1), fontsize=14)

plt.yticks(fontsize=14)
# 添加图形信息
plt.xlabel("Rating Score Margin", fontsize=18)
plt.ylabel(target_metric_name, fontsize=18)
plt.legend(fontsize=18)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()


# embedding_distance_relationship
# reward_score_diff
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datasets import load_dataset

# 载入数据
dataset = load_dataset("json", data_files="ultrafeedback_with_learning_order.json")['train']

# 模型名称
# model_names = ['gemma', 'llama', 'mistral', 'qwen-2.5-7b']
model_names = ['llama', 'mistral', 'qwen-2.5-7b']
show_errorbar=False
# 创建图形
plt.figure(figsize=(6, 4))
all_score_diffs = []

# 遍历每个模型
for model_name in model_names:
    # 过滤掉 None 值
    filtered_data = [
        (x["score_diff"], x[f"{model_name}_learning_order"])
        for x in dataset
        if x["score_diff"] is not None and x[f"{model_name}_learning_order"] is not None
    ]

    df = pd.DataFrame(filtered_data, columns=["score_diff", "learning_order"])
    
    # 四舍五入 score_diff 到最近 0.5
    df["score_diff"] = df["score_diff"].apply(lambda x: round(x * 2) / 2)

    # 计算 mean 和 std
    df_grouped = df.groupby("score_diff")["learning_order"].agg(["mean", "std"]).reset_index()
    
    all_score_diffs.extend(df_grouped["score_diff"].tolist())

    if show_errorbar:
        plt.errorbar(
            df_grouped["score_diff"],
            df_grouped["mean"],
            yerr=df_grouped["std"],
            fmt='-o',
            label=model_name,
            capsize=3,
            alpha=0.8,
            linewidth=2,
        )
    else:
        plt.plot(
            df_grouped["score_diff"],
            df_grouped["mean"],
            marker='o',
            label=model_name,
            alpha=0.9,
            linewidth=2,
        )


# 设置 x 轴刻度
min_x = np.floor(min(all_score_diffs))
max_x = np.ceil(max(all_score_diffs))
plt.xticks(np.arange(min_x, max_x + 0.5, 1), fontsize=14)
plt.yticks(fontsize=14)

# 添加图形信息
plt.xlabel("Raw Score Margin", fontsize=18)
plt.ylabel("Validation DPO loss", fontsize=18)
# plt.title("Mean Learning Order vs Score Difference (with Std)", fontsize=14)
plt.legend(fontsize=16)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

# dpo_loss_relationship
```
